```python
cols = ['oil_pct_chg', 'gold_pct_chg', 'set_pct_chg']
colors = ['.6', 'gold', 'c']
ax = df[cols].plot(kind='line', color=colors, figsize=(12, 5));
ax.axhline(0, color='green', linestyle='--', linewidth=1)
```

```python
ax = df.loc[::12,cols].plot(kind='line', color=colors)
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}%".format(v * 100)))
```

```python
ax = df.loc['2017-01-31':,cols].plot(kind='line', color=colors)
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}%".format(v * 100)))
```

```python
df[cols].plot(kind='line', subplots=True, color=colors, figsize=(12, 4), layout=(1, 3), sharey=True);
```

```python
base_date = '2003-01-31'
df['oil_pct_diff'] = df.Oil / df.Oil.at[base_date] - 1
df['gold_pct_diff'] = df.Gold / df.Gold.at[base_date] - 1
df['set_pct_diff'] = df.SET / df.SET.at[base_date] - 1
df
```

```python
ax  = df.loc[base_date:, ['oil_pct_diff', 'gold_pct_diff', 'set_pct_diff']].plot(kind='line', figsize=(10, 5), color=colors)
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}%".format(v * 100)))
```

```python
fig, ax = plt.subplots(1, 3, sharex=False, sharey=False, figsize=(12, 4))
s = ['Oil', 'Gold', 'SET']
title = ['OPEC Oil price US$/barrel', 'Gold US$/troy ounce', 'SET index']
color = ['.4', 'gold', 'c']
plt.tight_layout()
for i in range(len(s)):
    ax0 = df.loc['2003-01-31':, s[i]].plot(kind='line', color=color[i], ax=ax[i], title=title[i])
    ax0.axhline(df[s[i]].mean(), color='.6', linestyle='--', linewidth=1)
    ax0.set_xlabel('')
    ax0.set_ylabel('')
    ax0.yaxis.set_major_formatter(FuncFormatter(lambda v, p: "{:,.0f}".format(v)))
```

# Lambda function

```python
t = [0, 10, 20, 30, 40, 50]
```

```python
f = map(lambda c: c * 9 / 5 + 32, t)
list(f)
```

```python
def c2f(c):
    return (c * (9 / 5)) + 32
```

```python
f2 = map(lambda c: c2f(c), t)
list(f2)
```

```python
df = pd.DataFrame({'celsius': t})
df
```

```python
%timeit -n1000 df.celsius * 9/5+32
```

```python
%timeit -n1000 df.celsius.map(lambda c: c*9/5+32) #faster in one series
```

```python
%timeit -n1000 df.celsius.apply(lambda c: c*9/5+32)
```
